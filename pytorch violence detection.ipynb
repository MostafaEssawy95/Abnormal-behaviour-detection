```python
from torchvision.models import vgg16_bn
```

```python
class violent_net(nn.Module):
    def __init__(self, pretrained=True):
            super().__init__()
            self.encoder = vgg16_bn(pretrained=pretrained).features  
            self.convblock=nn.Sequential(*self.encoder[:])
            self.flatten=nn.Flatten()
            self.lstm=nn.LSTM(input_size=4608, hidden_size=20, num_layers=2,bidirectional=True)
            self.flatten=nn.Flatten()
            self.linear1=nn.Linear(400,32)
            self.activation1=nn.ReLU()
            self.linear2=nn.Linear(32,16)
            self.activation2=nn.ReLU()
            self.linear3=nn.Linear(16,2)
            self.activation3=nn.Softmax(dim=1)
            
    def forward(self, x1,x2,x3,x4,x5,x6,x7,x8,x9,x10):
        x1 = self.convblock(x1)
        x1=self.flatten(x1)
        x2 = self.convblock(x2)
        x2=self.flatten(x2)
        x3 = self.convblock(x3)
        x3=self.flatten(x3)
        x4 = self.convblock(x4)
        x4=self.flatten(x4)
        x5 = self.convblock(x5)
        x5=self.flatten(x5)
        x6 = self.convblock(x6)
        x6=self.flatten(x6)
        x7 = self.convblock(x7)
        x7=self.flatten(x7)
        x8 = self.convblock(x8)
        x8=self.flatten(x8)
        x9 = self.convblock(x9)
        x9=self.flatten(x9)
        x10 = self.convblock(x10)
        x10=self.flatten(x10)
        x=torch.cat((x1,x2,x3,x4,x5,x6,x7,x8,x9,x10),0)
        x=torch.unsqueeze(x,1)
        x=self.lstm(x)
        x=torch.cat((x[0][0],x[0][1],x[0][2],x[0][3],x[0][4],x[0][5],x[0][6],x[0][7],x[0][8],x[0][9]),1)
        x=self.flatten(x)
        x=self.linear1(x)
        x=self.activation1(x)
        x=self.linear2(x)
        x=self.activation2(x)
        x=self.linear3(x)
        #x=self.activation3(x)
        
        
       # return x1,x2,x3,x4,x5,x6,x7,x8,x9,x10
        return x
```

```python
ce = nn.CrossEntropyLoss()
def violenceLoss(preds, targets):
    ce_loss = ce(preds, torch.max(targets,1)[0])
    acc = (torch.max(preds, 1)[1] == targets).float().mean()
    return ce_loss, acc
```

```python
def train_batch(model, data, optimizer, criterion):
    model.train()
    ims1,ims2,ims3,ims4,ims5,ims6,ims7,ims8,ims9,ims10, ce_masks = data
    _masks = model(ims1,ims2,ims3,ims4,ims5,ims6,ims7,ims8,ims9,ims10)
    optimizer.zero_grad()
    loss, acc = criterion(_masks, ce_masks)
    loss.backward()
    optimizer.step()
    return loss.item(), acc.item()

@torch.no_grad()
def validate_batch(model, data, criterion):
    model.eval()
    ims1,ims2,ims3,ims4,ims5,ims6,ims7,ims8,ims9,ims10, ce_masks = data
    _masks = model(ims1,ims2,ims3,ims4,ims5,ims6,ims7,ims8,ims9,ims10)
    loss, acc = criterion(_masks,ce_masks)
    return loss.item(), acc.item()
```

```python
#model = violent_net().to(device)
criterion = violenceLoss
optimizer =optim.SGD(model.parameters(), lr=1e-3)
n_epochs = 14
```

```python
train_losses=[]
train_accuracy=[]
val_losses=[]
val_accuracy=[]
log = Report(n_epochs)
for ex in range(n_epochs):
    N = len(trn_dl)
    for bx, data in enumerate(trn_dl):
        loss, acc = train_batch(model, data, optimizer, criterion)
        log.record(ex+(bx+1)/N, trn_loss=loss, trn_acc=acc, end='\r')
    train_losses.append(loss)
    train_accuracy.append(acc)
    N = len(val_dl)
    for bx, data in enumerate(val_dl):
        loss, acc = validate_batch(model, data, criterion)
        log.record(ex+(bx+1)/N, val_loss=loss, val_acc=acc, end='\r')
    val_losses.append(loss)
    val_accuracy.append(acc)    
        
    log.report_avgs(ex+1)
```
